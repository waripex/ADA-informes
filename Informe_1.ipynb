# Problema de Ordenamiento


Este es un problema clásico el cual ha sido estudiado desde hace más de 50 años. Por otro lado, la gran cantidad de aplicaciones que tienen los problemas de este tipo generan un interés tanto teórico como práctico. De igual forma, producto de los distintos escenarios en los que se ha buscado aplicar es que han surgido variaciones de este problema. 


En general, el problema de ordenamiento tiene la siguiente estructura:
 

  **Entrada:** secuencia de *n* números  $[a_1, a_2, ... ,a_n]$

  **Salida:** Permutación ordenada de la secuencia de entrada: $[a'_1, a'_2, ... ,a'_n]$, de forma que $a'_1 \leq a'_2 \leq ... \leq a'_n$


Como ya se mencionó, a raíz del árduo estudio dedicado a este problema es que surgieron variantes del mismo. En esta oportunidad, se hará un análisis del InsertionSort.

# 1.- InsertionSort

## 1.1 Código

El codigo expuesto a continuacion, corresponde a una implementacion del Insertion Sort.

```python
# funcion InsertionSort
from termcolor import colored

def InsertionSort(arreglo, verbose = False):
  n = len(arreglo)
  nVuelta = 0 # variable para indicar la pasada en la que esta
  sum = 0 # variable sumatoria de la cantidad de veces que retrocedio el algoritmo
  
  if verbose == True: print("input:",arreglo)

  for i in range(1,n):

    
    #orden indica si el array esta ordenado o no
    orden = True
    
    # variable auxiliar que contabiliza los retrocesos
    cont = 0

    key = arreglo[i]
    j = i-1

    #Para orden descendente, cambiar key < arreglo [j] por ">"
    while j >= 0 and key < arreglo[j]:

      if verbose == True:
        
        if cont == 0:
          nVuelta+=1
          print("\nvuelta nº",nVuelta)

        print(str(arreglo[:j+1])[1:-1],",",colored(f"{str(key)}", "red"),",",str(arreglo[j+2:])[1:-1])

      orden = False
      cont += 1
      arreglo[j+1] = arreglo [j]
      j = j-1
    
    arreglo[j+1] = key
    if j >= 0 : cont +=1

    if orden == False and verbose == True:
      print("\nAl terminar:")
      if i-cont == 0:
        # Esta condicion se agrego apra obtener un correcto formato de printeo
        print(colored(f"{str(key)}", "blue"),",", str(arreglo[j+2:])[1:-1])
      else:
        print(str(arreglo[:j+1])[1:-1],",", colored(f"{str(key)}", "blue"),",", str(arreglo[j+2:])[1:-1])
    

    sum += cont # acumula la cantidad de retrocesos que toma ordenar c/numero

  if verbose == True: 
    print("\nOutput:",arreglo)

  media = sum/n # calculo de la media de retrocesos

  return arreglo,media,sum

# Ejemplo
A = [4,3,5,2]
print("Entrada:",A)
A,m,c = InsertionSort(A,verbose=False)
print("Salida:",A)
print("# Comparaciones:",c)
print("Media de comparaciones:",m)
```

## 1.2 Descripcion del algoritmo

El algortimo InsertionSort recibe como entrada una lista o arreglo con *n* elementos que se desean ordenar. Al finalizar, los datos de la lista se ordenaran dentro de la misma, sin la necesidad de una variable auxiliar.

>Al comenzar, se considera que el primer elemento ( $a[0]$ )  del arreglo se encuentra ordenado y los datos desde $a[1]$ hasta $a[n]$  son los que se necesitan ordenar, por lo que, la comparacion entre los valores inicia con el elemento $a[1]$.

>Posteriormente, para cada elemento $a[i]$ se recorre el arreglo, desde la posicion 0 hasta $i-1$, con la intencion de encontrar la posicion correcta. Esta comparacion consiste en verificar si el elemento $a[i]$ es $<$ o $>$ que el elemento $a[i-1]$ o $a[i+1]$, segun sea el caso de orden ascendente o descendente, respectivamente.

>Este procedimiento se ejecutara para cada posicion del arreglo hasta la ultima de ellas, $(n-1)$.

Cuando la variable *verbose* es *True*, el algoritmo mostrara la informacion de lo que ocurre paso a paso durante la ejecucion de la funcion. Es importante notar que, el algoritmo solo mostrara el desarrollo de los elementos que esten desordenados, en los datos que estan ordenados solo contabilizara la comparacion hecha.

## 1.3 Ejemplo

En el siguiente ejemplo, se busca ordenar un arreglo de menor a mayor, mediante la metodo de ordenamiento *InsertionSort*:

Consideremos inicialmente el arreglo:
> $ a = [ 3 , \mathring{5} , 2 , 8 ]$

Como se menciono, el 3 se considera ordenado y se procede a posicionar el 5. Para hacerlo se plantea si  3 < 5, como es cierto no se realiza ningun cambio de posicion, por lo que, en la primera pasada el arreglo queda igual.

>$ a = [ 3 , 5 , \mathring{2} , 8 ]$

En la segunda pasada se verifica si 5 < 2, como no es cierto, se cambian de posicion de forma que:

>$ a = [ 3 , 5 \leftrightarrows 2 , 8 ] \Longrightarrow a = [ 3 , 2 , 5 , 8 ]$

Este procedimiento se vuelve a realizar, pero entre 3 y 2: 
>$ a = [ 3 \leftrightarrows 2 , 5 , 8 ] \Longrightarrow a = [ 2 , 3 , 5 , 8 ]$

Finalmente, se comprueba si 5 < 8, como es cierto no se realiza ningun cambio y el programa finaliza entregando el arreglo ordenado.
> $ \therefore a = [ 2 , 3 ,5 , 8]$

# 2.- Tiempo de ejecucion 

###*Teorema ( Tiempo de ejecucion)*

*El algoritmo **InsertionSort** tiene un **tiempo de ejecucion** de $O(n^2)$ en el peor caso.*


### 2.1 **Prueba del Teorema**
Al observar el codigo, se puede apreciar que el peor caso del algoritmo ocurre cuando el menor elemento se encuentra al final del arreglo. Para este caso, en cada vuelta el menor elemento se moveria una casilla a la izquiera, por lo que, se necesitarian $(n-1)$ vueltas para que los elementos quedaran ordenados al finalizar el algortimo.

### 2.2 Ejemplo

```python
a =[2,3,4,1];
print("Entrada no ordenada:",a)
a,m,c=InsertionSort(a,True)
print("Salida ordenada:",a)
print("Total de comparaciones realizadas:",c)
```

En cada iteracion, en caso de estar desordenado, el algoritmo ira moviendo a la izquierda el elemento a ordenar, representado en rojo. En este caso, solo el 1 estaba fuera de posicion, por lo que realizo 3 comparaciones. Contabilizando la comparacion entre 2-3 y 3-4, genera un total de 5 comparaciones.

En el caso general, se deberian realizar $\sum_{i=0}^{n-1}i = \frac{n(n-1)}{2}$ comparaciones. Por lo que, **el tiempo de ejecucion del algoritmo, en el peor caso** es $O(n^2).\square$

# 3.- Correctitud

### **Teorema (correctitud)**

*El algoritmo **InsertionSort** genera un arreglo $[a'_1, a'_2, ... , a'_n]$ con los mismos elementos del arreglo de entrada ordenados de menor a mayor, de forma que $ a'_1 \leq a'_2 \leq ... \leq a'_n$.*

## Prueba del Teorema

Previamente se indico que en cada iteracion del bucle mas grande del InsertionSort, el subarreglo $a[0...i-1]$ corresponde a los elementos ordenados y el resto del arreglo corresponde a los elementos por ordenar.

Por lo que probaremos la sigueinte **propiedad invariante de bucle** para el *InsertionSort*:

> Al comienzo de cada iteracion del bucle mas grande, el subarreglo $a[0 ... i-1]$ contiene los mismos elementos que el arreglo original, pero ordenados de menor a mayor.

### **Inicializacion** 

Al comienzo de la primera iteracion, $i$ inicia con el valor 1, esto implica que,  $a[0...i-1]$ corresponde al primer elemento del arreglo original: $a[0]$ y ademas, se encuentra trivialmente ordenado.

### **Mantencion**

Se asume que antes de la comenzar la iteracion, el arreglo $a[0...i-1]$ se encuentra ordenado y contiene los elementos del arreglo original.

Durante la iteracion, se desplazara el nuevo elemento desde la posicion $i$ hasta su posicion correcta. Por lo tanto, despues de la iteracion, el arreglo $a[0...i-1]$ tambien contendra los elementos del arreglo original y se mantendra ordenado, porque se agrego el $i-esimo$ elemento.

### **Correctitud**

Finalmente, como la propiedad invariante de bucle es verdadera al inicio del bucle y se mantiene en cada iteracion, podemos asegurar que al finalizar la $n-esima$ iteracion del algoritmo, se genera un arreglo $[a'_1, a'_2, ... , a'_n]$, con los mismos elementos del arreglo de entrada y ordenados de menor a mayor. $\square$

## 4.2 Analisis del algoritmo con un arreglo casi completamente ordenado

En esta seccion se tomaran las mismas consideraciones del experimento 4.1, con la salvedad que el arreglo a graficar sera entregado ordenado casi en su totalidad.

```python
x=[n for n in range(5,30)] 
y1=[(n*(n-1))/2 for n in range(5,30)] # peor caso
y2=[n-1 for n in range(5,30)] # mejor caso
y=[]; 

for n in range(5,30):

  # arreglo casi ordenado
  a=[]
  for i in range(1,n+1):
    a.append(i)
  
  # se intercambia el 1 por el elemento n-1
  a[0] = a[n-1]
  a[n-1] = 1

  a,m,cont = InsertionSort(a,False)
  y.append(cont)

plt.plot(x,y)
plt.plot(x,y1)
plt.plot(x,y2)
plt.legend(["Insertion Sort", "peor caso teorico", "mejor caso teorico"])

plt.xlabel('n')
plt.ylabel('numero de comparaciones')
plt.show()
```

### BubbleSort

Para la siguiente grafica se incluira el codigo del *BubbleSort* y se expresara el tiempo de ejecucion del algoritmo en funcion del tamaño del problema par arreglos generados aleatoriamente de tamaños entre 5 y 200.

```python
# Bubble Sort
import random
from termcolor import colored
import copy

def bubble_sort(a, verbose=False):
    n = len(a)
    T = 0 #contador de comparaciones

    if verbose == True: print("input array:",a)

    for i in range(1,n+1):
        if verbose == True: print("\nPasada",i)
        # Flag que indica si el arreglo ya se encuentra ordenado
        already_sorted = True

        # Se recorre el arreglo
        for j in range(n - i):
            T +=1
            if a[j] > a[j + 1]:
                # Si el elemento es mayor al siguiente se intercambia
                a[j], a[j + 1] = a[j + 1], a[j]
                if verbose == True: 
                  print(str(a[:j])[1:-1],",",colored(f"{str(a[j:j+2])[1:-1]}","red"),",",str(a[j+2:])[1:-1])

                already_sorted = False

        if already_sorted == False and verbose == True: 
           print("\nAl finalizar pasada:")
           print(str(a[:n-i])[1:-1],",",colored(f"{str(a[n-i:])[1:-1]}","blue"))

        if already_sorted:
            if verbose == True: print("is sorted!")
            break

    if verbose == True: print("\noutput array:",a)

    return a, T
```

```python
# Grafica del tiempo de ejecucion en funcion del tamaño del arreglo

x=[]; y=[]

for n in range(5,200):

  a = random.sample(range(1, 1000), n)

  #la función repeat está ejecutando 20 veces bubble_sort con el arreglo a y retorna el mejor tiepo obtenido.
  t = repeat(setup="from __main__ import bubble_sort", stmt=f"bubble_sort({a})", repeat=1, number=10)

  x.append(n)
  y.append(t)


plt.plot(x,y)
plt.legend(["BubbleSort"])

plt.xlabel('n')
plt.ylabel('tiempo ms')
plt.show()
```
